```python
sw.dropna(subset=['name','rating', 'rating_count', 'cost', 'cuisine','address'])
```

```python
m=sw['lic_no'].mode()
sw.fillna(m)

#Filtering the null rows from null valued columns
sw = sw[-sw["name"].isnull()]
sw = sw[-sw["rating"].isnull()]
sw = sw[-sw["rating_count"].isnull()]
sw= sw[-sw["cost"].isnull()]
sw = sw[-sw["cuisine"].isnull()]
sw = sw[-sw["address"].isnull()]
sw = sw[-sw["lic_no"].isnull()]

round(100*(sw.isnull().sum()/len(sw.index)), 2) 
```

```python
sw['rating_count'].value_counts()
```

```python
#converting very few ratings to 1
def fun(n):
    try:
        num = n.split(" ")[0]
        if num.split('+')[0].isdigit():
            return float(num.split('+')[0]) 
        else:
            return 1 
    except:
        return n 

sw['rating_count']=sw['rating_count'].apply(fun) 
```

```python
#removing ruppe sign and converting cost to float

def fun(a):
    try:
        return float(a.split('₹')[1])
    except:
        return 0
sw['cost']=sw['cost'].apply(fun)
```
